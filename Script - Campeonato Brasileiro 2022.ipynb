```python
#Importando as bibliotecas
import pandas as pd
```

```python
#Colocando na variável os dados do site 
url = 'https://www.cnnbrasil.com.br/esporte/futebol/tabela-brasileirao-serie-a-2022/'
```

```python
#Leitura da variável e imprimindo os valores
html = pd.read_html(url)
html
```

```python
#Transformando a lista em um Data Frame
df_html = pd.read_html('https://www.cnnbrasil.com.br/esporte/futebol/tabela-brasileirao-serie-a-2022/')[0]
display(df_html)
```

```python
#Inserindo no DataFrame os Estados(UF) de cada time
df_html.insert(1, 'estado', ['SP','RS','RJ','SP','RJ','PR','MG','CE','SP','MG','RJ','SP','GO','SP','PR','MT','CE','GO','SC','RS'], True)
display(df_html)
```

```python
#Inserindo no DataFrame a quantidade de títulos brasileiros de cada time
df_html.insert(10, 'titulos', [11,3,4,7,8,1,2,0,6,0,2,8,0,0,1,0,0,0,0,0], True)
display(df_html)
```

```python
#Transformando os dados para o Excel
arquivo = 'tabela_criada.xlsx'
df_html.to_excel(arquivo)
```
